What is the percentage of attrition in the age category between 28 years and 38 years?

```python
# Joint probability- Probabilty of an employee's age group between 28-38 and got churned from the company

def age_joint(age):
    age_bivariate=pd.crosstab(age,columns=att.Attrition)
    return round(age_bivariate/len(att)*100,2)

age_joint((att["Age"]>=28) & (att["Age"]<39))

# P(Employee age group between 28-38 and got churned from the company)=0.0782 i.e approximately 8%
```
